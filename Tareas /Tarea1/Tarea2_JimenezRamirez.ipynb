### 13 
Escribe el código para sacar el "mean absolute error" de total bill

$$MAE =  \lvert \frac{x - E[X]}{n} \lvert$$


comparalo con si simplemente agarras ```np.abs()``` de toda la fila



````python
lista = [12 ,13, 50, 60]
import numpy as np 
promedio = np.mean(lista)
promedio
````

````python
lista_diferencias = []
for i in lista:
    diferencia = abs(i - promedio)
    lista_diferencias.append(diferencia)

np.mean(lista_diferencias)
````
