# Analisando otra fuente de datos datos 

```python
df = pd.read_csv('weight-height.csv')
```

```python
df['Height'] = np.round(df.Height * 2.54) / 100
df['Weight'] = np.round(df.Weight * 0.453592)
```

### ¿Como es el peso en este dataset?

```python
df.Weight.describe()
```

### ¿Como es la altura en este dataset?

```python
df.Height.describe()
```

## Visualiza este dataset 

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(df.Weight, df.Height) 
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.title('Altura vs Peso')
plt.grid()
```

```python
#Saber el nombre de las columnas 
df.keys()
```

## ¿Hay diferencia en el genero?

```python
df.groupby('Gender').mean()
```

```python
colors = {'Female' : 'red', 'Male' : 'blue'}
plt.scatter(df.Weight, df.Height, c= df.Gender.map(colors), alpha = 1)
plt.xlabel('Peso')
plt.ylabel('Altura')
plt.title('Altura vs Peso')
plt.grid()
```

### Usando aggregate

```python
df.groupby('Gender').agg({'Weight' : ['mean', 'median', 'std'],
                          'Height' : ['mean', 'median', 'std']})
                          
```
